```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("houses.csv")
df.head()
```

```python
df['house_type'].value_counts().plot(kind="pie", title="Most popular house types in Poland")
plt.show()
```

```python
df['price'].value_counts().plot(kind="bar")
plt.show()
```

```python
df['construction_costs'].plot(kind="line")
plt.title("Value by Population")
plt.ylabel("construction_costs")
plt.xlabel("index")
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.pairplot(df, corner=True)
plt.savefig("seaborn.png")
```

```python
sns.pairplot(df, corner=True, kind="kde")
```

```python
sns.heatmap(df.corr().abs().round(1), square=True, annot=True)
```

```python
df.plot("price", "size", "scatter")
```
