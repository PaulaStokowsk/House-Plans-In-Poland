```python
import pandas as pd
import numpy as np
import time
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import matplotlib.pyplot as plt
import os
import logging

import re
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
# Go to the page
page_url = "https://www.archon.pl/projekty-domow/filtr/projekty-domow-jednorodzinnych/4509?utf8=%E2%9C%93&view=&p%5Bsort%5D=ca&p%5Bhaf%5D=&p%5Bhat%5D=&p%5Bfn%5D%5B%5D=&p%5Bhb%5D%5B%5D=&p%5Bsr%5D%5B%5D=&p%5Bhg%5D%5B%5D=&p%5Brn%5D%5B%5D=&p%5Bbat%5D=&p%5Bpsx%5D=&p%5Bpsy%5D="
driver.get(page_url)

driver.execute_script("window.scrollTo(0, 1080)") 

time.sleep(6)
```

```python
#Closing popup banner
driver.find_element(By.ID, value = "e_katalog_popup_banner_close").click()
```

```python
# List of elements from the website
titles = driver.find_elements(By.XPATH, value="//a[@class='product-impression blue-hover']")
house_types = driver.find_elements(By.XPATH, value="//div[@class='small max-w-300']")
prices = driver.find_elements(By.XPATH, value="//span[@class='text-line-through small text-muted']")
sizes = driver.find_elements(By.XPATH, value="//span[@class='c-help tooltip-modal-init']")
construction_costs = driver.find_elements(By.XPATH, value="//div[@class='total_house_cost_min']")

#Gathering data
houses = []

for i in range(len(titles)):
    if bool(re.search(r'\bdomu\b', sizes[i].get_attribute('title'))):
        houses.append({'title': titles[i].text, 
                       'house_type': house_types[i].text, 
                       'price': prices[i].text.replace(" zł", "").replace(" ", ""), 
                       'size': round(float(sizes[i].text.replace("+", "").replace(",", "."))), 
                       'construction_costs': re.findall(r'\d+ \d+', construction_costs[i].text)[0].replace(" ", "")})
```

```python
df = pd.DataFrame(houses)
df
```

```python
#Finding next page
element = driver.find_element(By.XPATH, value="//a[@title='Następna']")

while element.text == 'NASTĘPNA ›':
    
    # List of elements from the website
    titles = driver.find_elements(By.XPATH, value="//a[@class='product-impression blue-hover']")
    house_types = driver.find_elements(By.XPATH, value="//div[@class='small max-w-300']")
    prices = driver.find_elements(By.XPATH, value="//span[@class='text-line-through small text-muted']")
    sizes = driver.find_elements(By.XPATH, value="//span[@class='c-help tooltip-modal-init']")
    construction_costs = driver.find_elements(By.XPATH, value="//div[@class='total_house_cost_min']")

    #Gathering data
    for i in range(len(titles)):
        if bool(re.search(r'\bdomu\b', sizes[i].get_attribute('title'))):
            houses.append({'title': titles[i].text, 
                           'house_type': house_types[i].text, 
                           'price': prices[i].text.replace(" zł", "").replace(" ", ""), 
                           'size': round(float(sizes[i].text.replace("+", "").replace(",", "."))), 
                           'construction_costs': re.findall(r'\d+ \d+', construction_costs[i].text)[0].replace(" ", "")})

    driver.find_element(By.XPATH, value="//a[@title='Następna']").click()
```

```python
df = pd.DataFrame(houses)
df
```

```python
df.to_csv('houses.csv', index=False)
```
